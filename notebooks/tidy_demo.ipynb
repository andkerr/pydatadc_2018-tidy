```python
ebola = pd.read_csv('../data/country_timeseries.csv')
```

```python
ebola_long = pd.melt(
    ebola,
    id_vars=['Date', 'Day'],
)
```

```python
variable_split = ebola_long['variable'].str.split('_')
```

```python
variable_split.head()
```

```python
# ???
# equiv to variable_split.str[0]
variable_split.str.get(0)
```

```python
ebola_long['status'] = variable_split.str.get(0)
ebola_long['country'] = variable_split.str.get(1)
```

```python
ebola_long.head()
```

### String accessor split() can expand to DataFrames

```python
ebola_long[['status', 'country']] = (ebola_long['variable']
                                     .str
                                     .split('_', expand=True))
```

```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather.head()
```

```python
weather_melt = pd.melt(
    weather,
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```

```python
weather_tidy.reset_index().head()
```
